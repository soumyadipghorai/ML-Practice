#### Pipelining

```python
X = abalone_data.iloc[:, :-1]
y = abalone_data.iloc[:, -1]
```

```python
X[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
numeric_features = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']
categorical_features = ['Sex']
```

```python
numeric_transformer = Pipeline(steps= [
    ("imputer", SimpleImputer(
        missing_values=0, strategy="constant", fill_value=0.107996
    )),
    ("scaler", StandardScaler())
])

categorical_transformer= OneHotEncoder(handle_unknown="ignore")
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

```python
# build tree with max depth = 3
clf = Pipeline(steps = [
    ("preprocessor", preprocessor),
    ("classifier", DecisionTreeClassifier(max_depth=3, random_state=42))
])
```

```python
clf.fit(X_train, y_train)
print("model score: %0.3f" % clf.score(X_test, y_test))
```

```python
y_pred = clf.predict(X_test)
```

```python
comparison = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.values.reshape(len(y_test), 1)), 1)
for each in comparison:
    print(each)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix \n', cm)
```

```python
#  for older version
"""
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(clf, X_test, y_test, cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.show()
"""

# new version
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_test_predicted = clf.predict(X_test)
cm = confusion_matrix(y_test, y_test_predicted)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title('Confusion matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
CR = classification_report(y_test, y_pred)
print('Classification report \n')
print(CR)
```

Cross- validation

```python
from sklearn.model_selection import cross_val_score
acc = cross_val_score(estimator=clf, X = X_train, y= y_train, cv=10)
print(type(acc))
print('Accuracy of each fold', list(acc*100))
print("Accuracy: {:.2f} %".format(acc.mean()*100))
```

Visualising the decision tree

HP tuning

```python
X_train_new = preprocessor.fit_transform(X_train)

from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth': range(1,10),
                'min_samples_split': [2, 4, 6, 8, 10]}

scores = ['recall']
for score in scores :
    print()
    print(f'Tuning HPT for {score}')
    clf_CV = GridSearchCV(
        DecisionTreeClassifier(), param_grid=param_grid,
        scoring = f'{score}_macro'
    )

    clf_CV.fit(X_train_new, y_train)
    print('Best params: ', clf_CV.best_params_)
    print('Grid score: ')
    means = clf_CV.cv_results_['mean_test_score']
    stds = clf_CV.cv_results_['std_test_score']

    for mean, std, params in zip(means, stds, clf_CV.cv_results_['params']) :
        print(f'{mean: 0.03f} for (+/-{std*2:0.03f}) for {params}')
```

```python
# train with the best params
clf2 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(max_depth = 5, min_samples_split = 4, random_state = 42))
])

clf2.fit(X_train, y_train)
print('model score : ', clf2.score(X_test, y_test))
```

### Iris Dataset

```python
from sklearn.datasets import load_iris
features, labels = load_iris(return_X_y = True, as_frame=True)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

dt_pipeline = Pipeline(steps=[
    ("scaling", MinMaxScaler()),
    ("clf", DecisionTreeClassifier(max_depth=3, random_state=42))
])
```

```python
dt_pipeline.fit(train_features, train_labels)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

ConfusionMatrixDisplay.from_estimator(dt_pipeline, test_features, test_labels);
```

```python
print(classification_report(test_labels, dt_pipeline.predict(test_features)))
```

```python
data = load_iris()
```

```python
from sklearn import tree
plt.figure(figsize=(20,8), facecolor="w")

a = tree.plot_tree(
    dt_pipeline[-1], feature_names = features.columns,
    class_names = data.target_names, rounded = True, filled = True
)

plt.show()
```

```python
tree_rules = tree.export_text(
    dt_pipeline[-1], feature_names = list(features.columns)
)

print(tree_rules)
```

```python
importance = pd.DataFrame({'feature': features.columns,
                            'importance': np.round(dt_pipeline[-1].feature_importances_, 3)})
importance.sort_values('importance', ascending=False, inplace=True)
print(importance)
```

```python
hp_grid = [{
    'clf__max_depth' : range(1, 6),
    'clf__min_samples_split' : range(2, 11, 2)
}]

gs_clf = GridSearchCV(dt_pipeline, hp_grid, scoring = 'f1_macro')
gs_clf.fit(train_features, train_labels)

print('best params')
print(gs_clf.best_params_)
print('Grid score: ')
means = gs_clf.cv_results_['mean_test_score']
stds = gs_clf.cv_results_['std_test_score']

for mean, std, params in zip(means, stds, gs_clf.cv_results_['params']) :
    print(f'{mean: 0.03f} for (+/-{std*2:0.03f}) for {params}')
```

```python
ConfusionMatrixDisplay.from_estimator(gs_clf.best_estimator_, test_features, test_labels);
```
