### Bagging 
**Estimators**
* BaggingClassifier 
* BaggingRegressor

both the estimators take the following parameters: 
* base_estimator => default None, for classification it takes Decision tree classifier and for regression it takes decision tree regressor 
* n_estimators => 10 
* max_sample => can be int or float. if its int then we take those many samples if its float we take `max_sample * total_sample`
* max_features => can be int or float. if its int then we take those many features if its float we take `max_features * total_features`
* bootstrap => True, whether samples are drawn with replacement
* bootstrap_features => False, whether features are drawn with replacement
* oob_score => False, whether to use out-of-bag samples to estimate generalization error.

### RandomForest 
**Estimators**
* RandomForestClassifier 
* RandomForestRegressor

parameter can be classified as decision tree parameters and bagging parameters. 

**bagging parameters**
* n_estimators => number of trees. for classification 10, regression 100.
* bootstrap 
    * True => bootstrapped samples are used.
    * False => whole dataset is used.
* oob_score => specifies whether to use oob samples for estimating generalization error. only available when bootstrap = True.
* max_sample => number of samples to draw while bootstrapping 
    * None => whole training data 
    * int => that many samples from training 
    * float => `max_sample * total_sample` samples from training data. value should be between [0,1]
* random_state => controls randomness of features and samples selected during bootstrap 
* max_features => number of features to be selected while splitting.
    * int => value specified 
    * float => `value x # features` 
    * auto => sqrt(#features)
    * sqrt => sqrt(#features)
    * log2 => log2(#features)
    * None => #features

**decision tree parameters**
* criterion 
    * gini => for classification 
    * squared_error => regression 
* max_depth => controls the depth of tree. default is None which means the tree will be grown until all leaf nodes are pure or until leaves contain less than `min_samples_splits` samples
* min_samples_splits => split internal node until they contain `min_samples_splits` samples.
    * int => value specified
    * float => `value x # features`
* min_impurity_decrease => a node will be split if it reduces impurity at least by the value specified.
* ccp_alpha => the complexity of tree can be controlled by this through minimal cost complexity pruning procedure.

**trained random forest estimators**
* estimators_ => member variable contrains a collection of fitted estimators 
* feature_importances_ => member variable contains a list of important features 

**training and inference for random forest** 
* fit => builds forest of trees 
* decision_path => returns decision path 
* predict => returns class label in classification and output value in regression 
* predict_proba => returns prob for classification setup 
* predict_log_proba => returns log prob for classification setup
